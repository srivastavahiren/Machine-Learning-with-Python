##Importing libraries
    

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

##Importing dataset

```python
dataset = pd.read_csv('Mall_Customers.csv')
x = dataset.iloc[:,[3,4]].values
```

##Using Dendograms to find the optimal numbers of clusters


```python
import scipy.cluster.hierarchy as sch
dendogram = sch.dendrogram(sch.linkage(x,method='ward'))
plt.title("Dendogram")
plt.xlabel("Observation Points")
plt.ylabel("Euclidean Distance")
plt.show()
```

##Training the Hierachical Clustering model on the dataset

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters =5 , affinity = "euclidean", linkage="ward")
y_pred_hc = hc.fit_predict(x)
```

```python
print(y_pred_hc)
```

##Visualising clusters

```python
plt.scatter(x[y_pred_hc==0,0],x[y_pred_hc==0,1],s=100,c='blue',label ='Cluster1')
plt.scatter(x[y_pred_hc==1,0],x[y_pred_hc==1,1],s=100,c='red',label ='Cluster2')
plt.scatter(x[y_pred_hc==2,0],x[y_pred_hc==2,1],s=100,c='cyan',label ='Cluster3')
plt.scatter(x[y_pred_hc==3,0],x[y_pred_hc==3,1],s=100,c='magenta',label ='Cluster4')
plt.scatter(x[y_pred_hc==4,0],x[y_pred_hc==4,1],s=100,c='yellow',label ='Cluster5')
plt.title("Clusters of Customers")
plt.xlabel("Annual_income(k$)")
plt.ylabel("Spending Score(1-100)")
plt.show()
```
